# 3.Create appropriate graphs to visually represent the relationship between financial services accessed (non, momible, both) and age, gender, marital status, land ownership and type of income.
# MM = mobile money

```python
import seaborn as sns
import matplotlib.pyplot as plt    
sns.violinplot(x=money_df['mobile_money_classification'],y=money_df['Q1'])    
plt.title('Mobile Money Classification vs Age')
plt.xlabel('Mobile Money Classification ')
plt.xticks(np.arange(5), ('No Services', '  No MM/One other', 'MM only', 'MM/one other'))
plt.ylabel('Age')
plt.show()
```

From this graph we can determine that the age group with the most amount of people accesing financial services are
from 20 to 40. With mobile money and one other service holding the most people in that demographic and no services
the least.

```python
sns.countplot(data=money_df,
                  x = 'mobile_money_classification',
                  hue = 'Q2')
plt.title('Mobile Money Classification vs Gender')
plt.xlabel('Mobile Money Classification ')
plt.xticks(np.arange(5), ('No Services', '  No MM/One other', 'MM only', 'MM/one other'))
plt.ylabel('Number of People')
plt.legend(['Male', 'Female'])
plt.show()
```

From this graph we can determine that the women are less likely to use financial services. As you can see they are the majority in the categories No Services and No mobile money.

```python
sns.countplot(data=money_df,
                  x = 'mobile_money_classification',
                  hue = 'Q3')
plt.title('Mobile Money Classification vs Marital Status')
plt.xlabel('Mobile Money Classification ')
plt.xticks(np.arange(5), ('No Services', '  No MM/One other', 'MM only', 'MM/one other'))
plt.ylabel('Number of People')
plt.legend(['Married', 'Divorced', 'Widowed', 'Single/never married'])
plt.show()
```

Here we can see that people who are married are more likely to use financial services and that they are the largest part of the sample set. 

```python
sns.countplot(data=money_df,
                  x = 'mobile_money_classification',
                  hue = 'Q6')
plt.title('Mobile Money Classification vs Land Ownership')
plt.xlabel('Mobile Money Classification ')
plt.xticks(np.arange(5), ('No Services', '   No MM/One other', 'MM only', 'MM/one other'))
plt.ylabel('Number of People')
plt.legend(['Own Land', "Don't Own Land"])
plt.show()
```

From this graph we can see that many of the people who own land use mobile money. However, majority of the mobile money client base are people who don't own land.

```python
sns.countplot(data=money_df,
                  x = 'mobile_money_classification',
                  hue = 'Q8_1')
plt.title('Mobile Money Classification vs Income')
plt.xlabel('Mobile Money Classification ')
plt.xticks(np.arange(5), ('No Services', '  No MM/One other', 'MM only', 'MM/one other'))
plt.ylabel('Number of People')
plt.legend(['Other', 'Salary/Wages'])
plt.show()
```

Majority of the set we are working with have other sources of income. Therefore, majority of the mobile money customer base is made up of the other. But from people who earn a salary/wages most use the app.

# 4. Create appropriate graphs to visually represent the relationship between how often mobile services are used and age, gender, marital status, land ownership and type of income.

```python
sns.violinplot(x=money_df['Q16'],y=money_df['Q1'])    
plt.title('How often they used app in past 12 Months vs Age')
plt.xlabel('How often they used app in past 12 Months ')
plt.xticks(np.arange(5), ('Never', 'Daily', 'Weekly', 'Monthly', '             Less often than monthly'))
plt.ylabel('Age')
plt.show()
```

The older someone is the less likely they are to use the mobile money app. The most frequent users are between the age group 20-40

```python
sns.countplot(data=money_df,
                  x = 'Q16',
                  hue = 'Q6')
plt.title('How often they used app in past 12 Months vs Land Ownership')
plt.xticks(np.arange(5), ('Never', 'Daily', 'Weekly', 'Monthly', '             Less often than monthly'))
plt.xlabel('Mobile Money Classification ')
plt.ylabel('Number of People')
plt.legend(['Own Land', "Don't own land"])
plt.show()
```

As determined most of the people who do not own land make up majority of the mobile money customer base. Therefore, in this graph it makes sense that people who don't own land are the most frequent users. 

```python
sns.countplot(data=money_df,
                  x = 'Q16',
                  hue = 'Q2')
plt.title('How often they used app in past 12 Months vs Gender')
plt.xticks(np.arange(5), ('Never', 'Daily', 'Weekly', 'Monthly', '             Less often than monthly'))
plt.xlabel('Mobile Money Classification ')
plt.ylabel('Number of People')
plt.legend(['Male', 'Female'])

plt.show()
```

From this we can see that females use mobile money less frequently than men. This could be due to the fact that women overall use less financial services.

```python
sns.countplot(data=money_df,
                  x = 'Q16',
                  hue = 'Q3')
plt.title('How often they used app in past 12 Months vs Marital Status')
plt.xticks(np.arange(5), ('Never', 'Daily', 'Weekly', 'Monthly', '             Less often than monthly'))
plt.xlabel('Mobile Money Classification ')
plt.ylabel('Number of People')
plt.legend(['Married', 'Divorced', 'Widowed', 'Single/never married'])

plt.show()
```

People who are married are more likely to use mobile money whereas widowed people are least likely

```python
sns.countplot(data=money_df,
                  x = 'Q16',
                  hue = 'Q8_1')
plt.title('How often they used app in past 12 Months vs Income')
plt.xticks(np.arange(5), ('Never', 'Daily', 'Weekly', 'Monthly', '             Less often than monthly'))
plt.xlabel('Mobile Money Classification ')
plt.ylabel('Number of People')
plt.legend(['Other Types of Income', 'Salary/Wages'])
plt.show()
```

People with other sources of income are the most frequent users
